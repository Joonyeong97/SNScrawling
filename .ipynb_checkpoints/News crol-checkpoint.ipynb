```python
import requests
from bs4 import BeautifulSoup
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import nltk
from konlpy.tag import Twitter
import numpy as np
from PIL import Image
```

```python
from ckonlpy.tag import Twitter
text2 = []
t = Twitter()
t.add_dictionary(['트와이스', 'kf94','KF94','Kf94','kF94','타임라인','확진자','예방수칙','코로나19','corona19','Corona19',
                 '개소리','판매','제품','쿠팡','kf94마스크','KF94마스크','Kf94마스크','kF94마스크',
                 '우한폐렴','신종코로나','신종코로나바이러스','coronavirus','Coronavirus','사재기',
                 '복지부장관', '바이러스', '피해복구', '이만희', '문재인', '이재갑','한림대',
                 '감염내과', '교수님', '정치인', '입국금지','대변인', '청와대'], 'Noun')
```

```python
htt = 'https://news.naver.com/main/read.nhn?mode=LSD&mid=shm&sid1=100&oid=014&aid=0004380386'
htt2 = 'https://news.naver.com/main/read.nhn?mode=LSD&mid=shm&sid1=100&oid=022&aid=0003442433'
htt3 = 'https://news.naver.com/main/read.nhn?mode=LSD&mid=shm&sid1=102&oid=081&aid=0003069212'
htt4 = 'https://news.naver.com/main/read.nhn?mode=LSD&mid=shm&sid1=102&oid=047&aid=0002258682'
htt5 = 'https://news.naver.com/main/read.nhn?mode=LSD&mid=shm&sid1=105&oid=008&aid=0004365697'

http = [htt,htt2,htt3,htt4,htt5]
```

```python
for i in range(len(http)):
    res = requests.get(http[i])
    soup = BeautifulSoup(res.content, 'html.parser')
    body = soup.select('._article_body_contents')[0]
    text = body.get_text() # for p in body.find_all('div'))
    text2.insert(-1,text)
```

```python
import nltk
from konlpy.tag import Twitter
import numpy as np
tokens_ko = []

for i in range(len(text2)):
    tokens_ko.append(t.nouns(text2[i]))
```

```python
final = []
for _,q in enumerate(tokens_ko):
    for i in range(len(q)):
        final.insert(-1,q[i])
```

```python
ko = nltk.Text(final, name="첫번째")
data = ko.vocab().most_common(1000)
```

```python
import requests
from bs4 import BeautifulSoup
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import nltk
from konlpy.tag import Twitter
import numpy as np
from PIL import Image

tmp_data = dict(data)

wordcloud = WordCloud( font_path = '/Library/Fonts/NanumMyeongjo.ttf',
                       background_color='white').generate_from_frequencies(tmp_data)
plt.figure(figsize=(16,8))
plt.imshow(wordcloud)
plt.axis('off')
plt.show()
```
